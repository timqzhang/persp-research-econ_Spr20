```python
def corr_ind_process_wgi (title, abbr, df2):
    
    start_year=1996
    end_year=2018
    
    df=df2.copy()
    
    df=df.rename(columns={"Country/Territory": "Country Name", "Code": "Country Code"})
    df.insert(2, "Indicator Name", [title]*len(df), True)
    df.insert(3, "Indicator Code", [abbr]*len(df), True)
    df.insert(5, "1997", [np.NaN]*len(df), True)
    df.insert(7, "1999", [np.NaN]*len(df), True)
    df.insert(9, "2001", [np.NaN]*len(df), True)
    
    for i in range(1996,2019):
        df=df.rename(columns={i:str(i)})
    
    #print(df)

    for row in df.iterrows():
        #print(wdi_df_filter3.at[row[0],'1996'])
        data_list=[]
        t=True
        for year in range(start_year, end_year+1):
        
            if np.isnan(row[1][str(year)]):
                t=False
            data_list.append(row[1][str(year)])
        #print(t,data_list)
        if t==False:
            year_list=np.arange(start_year,end_year+1,1)
            for i,element in enumerate(data_list):
                if np.isnan(element):
                    data_list[i]=-100
                    year_list[i]=-100
                    
            if data_list.count(-100)>int(len(data_list)*2/3):
                df=df.drop(row[0])
                
            else:
                
                year_list = list(filter(lambda x: x != -100, year_list))
                data_list = list(filter(lambda x: x != -100, data_list))

                slope, intercept, r_value, p_value, std_err = linregress(year_list, data_list)
                #print(slope,intercept)
                for year in range(start_year, end_year+1):
                    if np.isnan(row[1][str(year)]):
                        df.at[row[0], str(year)]=(int(year)*slope+intercept)

    return df
```

```python
for key, value in indicator_name_dict.items():
    if 'WGI' not in key and 'ICRG' not in key:
        
        if key=='NY.GDP.PCAP.KD':
            panel=panel.replace({'Indicator Code': key}, 'WDI.GDP.PCAP.KD')
            indicator_name_dict['WDI.GDP.PCAP.KD']=value
            del indicator_name_dict[key]
        elif key=='NE.TRD.GNFS.ZS':
            panel=panel.replace({'Indicator Code': key}, 'WDI.TRADE.PER.GDP')
            indicator_name_dict['WDI.TRADE.PER.GDP']=value
            del indicator_name_dict[key]
        elif key=='NE.CON.GOVT.ZS':
            panel=panel.replace({'Indicator Code': key}, 'WDI.GOV.CONSUM.PER.GDP')
            indicator_name_dict['WDI.GOV.CONSUM.PER.GDP']=value
            del indicator_name_dict[key]
        elif key=='NY.GDP.DEFL.KD.ZG':
            panel=panel.replace({'Indicator Code': key}, 'WDI.INFLATION.KD.PER')
            indicator_name_dict['WDI.INFLATION.KD.PER']=value
            del indicator_name_dict[key]
        elif key=='BX.KLT.DINV.WD.GD.ZS':
            panel=panel.replace({'Indicator Code': key}, 'WDI.NET.INFLOW.FDI.PER.GDP')
            indicator_name_dict['WDI.NET.INFLOW.FDI.PER.GDP']=value
            del indicator_name_dict[key]
        elif key=='NY.GDP.TOTL.RT.ZS':
            panel=panel.replace({'Indicator Code': key}, 'WDI.NATUR.RESOUR.PER.GDP')
            indicator_name_dict['WDI.NATUR.RESOUR.PER.GDP']=value
            del indicator_name_dict[key]
        elif key=='NE.GDI.FTOT.ZS':
            panel=panel.replace({'Indicator Code': key}, 'WDI.CAPI.INVEST.PER.GDP')
            indicator_name_dict['WDI.CAPI.INVEST.PER.GDP']=value
            del indicator_name_dict[key]
```
